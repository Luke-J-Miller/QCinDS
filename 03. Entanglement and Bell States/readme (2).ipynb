# What is Quantum Entanglement?

## Definition:
A quantum state where two or more qubits are correlated such that the state of one cannot be described independently of the others
## Mathematical Form:
Non-separable state, e.g., $ |\psi\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle$
## Key Properties:
- Instantaneous correlations upon measurement
- No faster-than-light communication (no-signaling theorem)

## Fragile:
Decoherence destroys entanglement


## Non-Classical Nature:
Violates local realism; measurements on one affect the other "spookily" at a distance
## Relevance:
Core resource for quantum advantage in algorithms

# The EPR Paradox

## Historical Context:
Einstein, Podolsky, Rosen (1935) paper questioning quantum mechanics' completeness
## Thought Experiment:
- Two particles in entangled state, separated by large distance
- Measure position on one → instantly know position of the other
- Measure momentum on one → instantly know momentum of the other


## Issue:
Violates uncertainty principle if reality is local and deterministic
## Resolution:
Quantum mechanics is non-local but consistent; no hidden variables (per Bell)
## Implications for AI:
Entanglement enables distributed quantum computing for large-scale data processing

# Bell Inequalities

## John Bell (1964):
Testable inequalities to distinguish local hidden variable theories from quantum mechanics
## CHSH Inequality:
For local theories, correlation $ |E(a,b) + E(a,b') + E(a',b) - E(a',b')| \leq 2 $
## Quantum prediction:
Up to $ 2\sqrt{2} \approx 2.828 $


## Experimental Violations:
Aspect (1982), Weihs (1998), etc., confirm quantum mechanics
## Mathematical Setup:
Observables A, A' on first particle; B, B' on second
Expectation $ E = \langle \psi | A \otimes B | \psi \rangle $


## Significance:
Proves entanglement's non-classicality; basis for quantum cryptography

# Bell States (Maximally Entangled States)

## The Four Bell States:
$$
( |\Phi^+\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle) )\\
( |\Phi^-\rangle = \frac{1}{\sqrt{2}}(|00\rangle - |11\rangle) )\\
( |\Psi^+\rangle = \frac{1}{\sqrt{2}}(|01\rangle + |10\rangle) )\\
( |\Psi^-\rangle = \frac{1}{\sqrt{2}}(|01\rangle - |10\rangle) )\\
$$

## Properties:
Orthogonal basis for 2-qubit space; maximal entanglement (von Neumann entropy = 1)
## Measurement Correlations:
Perfect (anti-)correlation in certain bases
## In Quantum Computing:
Standard resource for protocols

# Creating Entangled States

## Circuit for Bell Pair $( |\Phi^+\rangle )$:
1. Start with $ |00\rangle $
2. Apply H to first qubit: $ \frac{1}{\sqrt{2}}(|00\rangle + |10\rangle) $
3. Apply CNOT (control first, target second): $ \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle) $


## General Method:
Superposition on control + controlled operation
## Variations:
Phase gates for other Bell states (e.g., X on second for $ |\Psi^+\rangle )$
## Theoretical Insight:
Entanglement generated by interactions (non-local gates)

# Measuring Correlations

## Joint Measurements:
### In computational basis:
For $ |\Phi^+\rangle $, outcomes 00 or 11 with 50% each; always same
### In Hadamard basis:
Apply H to both; correlations depend on state


## Correlation Function:
$ E(\theta) = \cos(2\theta) $ for angle between measurements
## Bell Test in Practice:
Measure in different bases to compute CHSH
## Visualization:
No Bloch for multi-qubit, but use statevector or density matrix
## Non-Classical Demo:
Classical simulation fails Bell inequality

# Applications: Quantum Teleportation

## Protocol Overview:
- Alice has qubit to send; shares Bell pair with Bob
- Alice measures her qubits (original + half-pair) in Bell basis
- Sends 2 classical bits to Bob
- Bob applies corrections (X/Z) to his half-pair → receives state


## Key Insight:
Transfers state without physical transport; uses entanglement
## Circuit:
Involves Bell creation, measurements, conditional gates
## Relevance to Data Science:
Secure data transfer in quantum networks for distributed ML

## Applications: Superdense Coding

### Protocol Overview:
- Alice sends 2 classical bits using 1 qubit via shared entanglement
- Alice applies Pauli gates (I, X, Z, XZ) to her half based on message
- Sends qubit to Bob; Bob measures in Bell basis to decode


### Advantage:
- Doubles classical channel capacity with entanglement
### Circuit:
Bell pair + Alice's operations + Bob's H and CNOT + measurements

## In AI Context:
Efficient communication in quantum-enhanced distributed systems

# Visualizing Entanglement

## Tools:
### Statevector:
List amplitudes (e.g., [0.707, 0, 0, 0.707] for $ |\Phi^+\rangle )$
### Density Matrix:
For partial tracing and entanglement measures
Hinton Diagram or City Plot in Qiskit for visualization


## Entanglement Measures:
Concurrence, negativity; for Bell states = 1 (max)
## Theoretical Example:
Compute reduced density matrix $ \rho_A = Tr_B(|\psi\rangle\langle\psi|) = \frac{1}{2}I $ (mixed state indicates entanglement)
