## **Section B – Short Problems** (4 × 5 pts = 20 pts)

**B1. (Bell-state verification)**
Prove that $\text{CNOT}(H\otimes I)|10⟩ = |\Psi^{+}\rangle$.  Then compute the probability of measuring outcome $|01⟩$ if an additional Z gate is accidentally applied to the first qubit **before** measurement.

---

**B2. (Query algorithm)**
For the Bernstein–Vazirani problem with secret string $s=1011$ ($n=4$), write the matrix representation of the oracle $O_s$ acting on the $n+1$ qubits (include the ancilla).  Demonstrate why one quantum query reveals $s$.

---

**B3. (Noise propagation)**
A single qubit experiences dephasing described by Kraus operators
$K_0=\sqrt{1-\lambda}\,I,\;K_1=\sqrt{\lambda}\,Z$.
Starting from $|+\rangle$, derive the output density matrix and compute the Bloch-vector length as a function of $\lambda$.

---

**B4. (QAOA energies)**
Consider a 3-vertex triangle graph with uniform weights (MaxCut).  Using depth-$p=1$ QAOA, the cost expectation is
$\langle H_C\rangle = \tfrac34+ \tfrac34\sin(4\gamma)\sin(2\beta) - \tfrac14\cos(4\gamma)$.
Find $\gamma$ and $\beta$ that maximise this expectation (give values modulo $\pi$).  What is the corresponding approximation ratio relative to the optimal cut value 3?
