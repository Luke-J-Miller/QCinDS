## **Section A – Multiple Choice** (10 × 2 pts = 20 pts)

*Circle the single best answer.*

1. A single application of the **T** gate on $|1⟩$ produces a relative phase of
- a) $e^{i\pi/2}$
- b) $e^{i\pi/4}$
- c) $-1$
- d) no phase change

2. Which pair of gates **always commutes**?
- a) $X$ and $Z$
- b) $Z$ and $T$
- c) $H$ and $X$
- d) CNOT(control = 0, target = 1) and $H$ on qubit 1

3. In a Bell-inequality (CHSH) experiment, a classical local-hidden-variable model is upper-bounded by
- a) 1
- b) 2
- c) $2\sqrt2$
- d) 4

4. The Bernstein–Vazirani oracle $\mathcal O_s$ acts on $|x⟩|b⟩$ as
- a) $|x⟩|b\oplus f(x)⟩$
- b) $(-1)^{s\cdot x}|x⟩|b⟩$
- c) $|x⟩|b⟩$ (identity)
- d) swaps the two registers

5. A **Toffoli** (CCX) gate can be decomposed into basic gates without ancillas using at minimum
- a) 1 CNOT and 2 $T$ gates
- b) 2-qubit gates only
- c) 6 CNOTs and single-qubit rotations
- d) it cannot be decomposed

6. In Grover’s algorithm the diffusion operator can be written as
- a) $\displaystyle 2|s\rangle\langle s|-I$
- b) $H^{\otimes n}$
- c) $\mathcal O_f$
- d) $X^{\otimes n}$

7. The **Pauli-twirling** approximation converts any noise channel into a …
- a) depolarising channel
- b) coherent over-rotation
- c) thermal relaxation channel
- d) read-out error

8. The depth of an $n$-qubit **inverse** QFT with nearest-neighbour connectivity scales as
- a) $O(n)$
- b) $O(n^2)$
- c) $O(2^n)$
- d) independent of $n$

9. In QAOA, increasing the depth parameter $p$ always …
-   a) guarantees higher approximation ratio
-   b) reduces circuit depth on hardware
-   c) increases parameter count $2p$
-   d) eliminates barren plateaus

10. The dominant cost of modular exponentiation in Shor’s algorithm scales with
- a) qubit count only
- b) gate count $O((\log N)^3)$
- c) number of oracle calls
- d) measurement shots

$\mathcal D_p(\rho)=(1-p)\rho+\tfrac{p}{3}(X\rho X+Y\rho Y+Z\rho Z)$.
Compute $\mathcal D_p(|+\rangle\langle+|)$ explicitly in matrix form.
